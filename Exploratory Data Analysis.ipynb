# EDA for Missing Pet Volume Data

```python
import pandas as pd
import duckdb

import plotly.express as px
```

```python
daily_report_counts_us = pd.read_csv('./data/daily_report_counts_us.csv', index_col='event_date', parse_dates=True)
daily_report_fig = px.line(daily_report_counts_us,
  x=daily_report_counts_us.index,
  y='reports',
  title='Daily Missing Pet Report Volume 2019 - 2022',
  width=720,
  labels={
    'reports': 'Report Volume',
    'event_date': 'Date Lost or Found'
  }
)
daily_report_fig.add_annotation(
  x='2020-12-27',
  y='255',
  text='New Year\'s Eve'
)
daily_report_fig.add_annotation(
  x='2020-07-03',
  y='507',
  text='4th of July'
)
```

```python
daily_report_counts_us['day_of_week'] = daily_report_counts_us.index.day_name()
fig = px.box(daily_report_counts_us,
  y='reports',
  x='day_of_week',
  points="all",
  color='day_of_week',
  width=720,
  title='Comparison of Report Volume Distributions by Day of Week',
  labels={
    'reports': 'Volume of Reports',
    'day_of_week': 'Day of Week'
  }
)
fig.show()
```
